```python
# prompt: create a student csv file with 25 records

import pandas as pd
import random

# Generate random data for 25 students
names = ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', 'Henry', 'Isabella', 'Jack',
         'Katherine', 'Liam', 'Mia', 'Noah', 'Olivia', 'Peter', 'Quinn', 'Ryan', 'Sophia', 'Thomas',
         'Ursula', 'Vincent', 'Wendy', 'Xavier', 'Yara']
ages = [random.randint(18, 25) for _ in range(25)]
majors = ['Computer Science', 'Mathematics', 'Physics', 'Biology', 'Chemistry', 'Engineering',
          'Economics', 'Psychology', 'History', 'English']
gpas = [round(random.uniform(2.5, 4.0), 2) for _ in range(25)]

# Create a DataFrame
df = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'Major': [random.choice(majors) for _ in range(25)],
    'GPA': gpas
})

# Save the DataFrame to a CSV file
df.to_csv('students.csv', index=False)
```

```python
# 3. Filter students with GPA greater than 3.5
high_gpa_students = df[df['GPA'] > 3.5]
print(high_gpa_students)
```

```python
# 4. Group students by major and calculate average GPA
average_gpa_by_major = df.groupby('Major')['GPA'].mean()
print(average_gpa_by_major)
```

```python
# 5. Sort students by age in descending order
sorted_by_age = df.sort_values('Age', ascending=False)
print(sorted_by_age)
```

```python
# Calculate the difference between each student's age and the average age
average_age = df['Age'].mean()
df['Age Difference'] = df['Age'].diff()  # Calculates difference between consecutive rows
print(df)
```

```python
# prompt: skewed distribution

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame

# Visualize the distribution of 'Age'
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()

# Visualize the distribution of 'GPA'
sns.histplot(df['GPA'], kde=True)
plt.title('Distribution of GPA')
plt.show()

# Calculate skewness
print("Skewness of Age:", df['Age'].skew())
print("Skewness of GPA:", df['GPA'].skew())

# You can also use box plots to visualize skewness
sns.boxplot(x=df['Age'])
plt.title('Box Plot of Age')
plt.show()

sns.boxplot(x=df['GPA'])
plt.title('Box Plot of GPA')
plt.show()
```

```python
# prompt: positive and negative skews

# Positive Skew Example
data_positive_skew = [1, 2, 3, 4, 100]  # Outlier on the right
sns.histplot(data_positive_skew, kde=True)
plt.title('Positive Skew')
plt.show()
print("Skewness:", pd.Series(data_positive_skew).skew())

# Negative Skew Example
data_negative_skew = [1, 10, 20, 30, 40]  # Outlier on the left
sns.histplot(data_negative_skew, kde=True)
plt.title('Negative Skew')
plt.show()
print("Skewness:", pd.Series(data_negative_skew).skew())
```

```python
import numpy as np

# Left Skew
left_skewed = np.random.beta(5, 1, 1000)
sns.histplot(left_skewed, kde=True)
plt.title('Left Skewed Distribution')
plt.show()
sns.boxplot(x=left_skewed)
plt.title('Box Plot of Left Skewed Data')
plt.show()



```

```python
# Symmetric
symmetric = np.random.normal(0, 1, 1000)
sns.histplot(symmetric, kde=True)
plt.title('Symmetric Distribution')
plt.show()
sns.boxplot(x=symmetric)
plt.title('Box Plot of Symmetric Data')
plt.show()
```

```python

# Right Skew
right_skewed = np.random.beta(1, 5, 1000)
sns.histplot(right_skewed, kde=True)
plt.title('Right Skewed Distribution')
plt.show()
sns.boxplot(x=right_skewed)
plt.title('Box Plot of Right Skewed Data')
plt.show()
```
